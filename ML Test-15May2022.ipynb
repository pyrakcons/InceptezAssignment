### 1. Cellphone Dataset

### NaiveBayes :

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
dfcellph = pd.read_csv('D:\Datascience\inceptez\Preprocessing\Cellphone.csv')
dfcellph
```

```python
dfcellph.shape
```

```python
dfcellph.columns
```

```python
x = dfcellph.drop('price_range',axis=1)
y = dfcellph['price_range']
```

```python
x
```

```python
y.value_counts()
```

```python
x.describe()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y,test_size=.25,random_state=2)
```

```python
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_prd = model.predict(x_test)
```

```python
acc_score = accuracy_score(y_test,y_prd)
```

```python
print(acc_score)
```

```python
confusion_matrix(y_test,y_prd)
```

```python
print(classification_report(y_test,y_prd))
```

#### Cell Phone Predict:

```python
cell_pred = model.predict([[563,1,0.5,1,2,1,41,0.9,145,5,6,1263,1716,2603,11,2,9,1,1,0]])
if cell_pred[0] == 0:
    print('LOW PRICE')
if cell_pred[0] == 1:
    print('MEDIUM PRICE')
if cell_pred[0] == 2:
    print('HIGH PRICE')
if cell_pred[0] == 3:
    print('VERY HIGH PRICE')        
print(cell_pred)     
```

 ### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
cell_pred = knn.predict([[563,1,0.5,1,2,1,41,0.9,145,5,6,1263,1716,2603,11,2,9,1,1,0]])
if cell_pred[0] == 0:
    print('LOW PRICE')
if cell_pred[0] == 1:
    print('MEDIUM PRICE')
if cell_pred[0] == 2:
    print('HIGH PRICE')
if cell_pred[0] == 3:
    print('VERY HIGH PRICE')        
print(cell_pred)  
```

### 2. Fraud Detection Dataset:

### Naive Bayes and KNN

```python
dffrddet = pd.read_csv('D:\Datascience\inceptez\Preprocessing\Fraud-Det.csv')
dffrddet
```

```python
dffrddet.describe()
```

```python
dffrddet['isFraud'].value_counts()
```

```python
dffrddet.columns
```

```python
dffrddet['type'].unique()
```

```python
dffrddet['type_label']= np.where(dffrddet['type']=='PAYMENT',1,
                                 (np.where(dffrddet['type']=='TRANSFER',2,(np.where(dffrddet['type']=='CASH_OUT',3,(np.where(dffrddet['type']=='CASH_IN',4,5)))))))
dffrddet
```

```python
dffrddet['type_label'].unique()
```

```python
x1= dffrddet[['step','type_label','amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']]
y1 = dffrddet['isFraud']
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x1, y1,test_size=.25,random_state=2)
```

```python
model1 = GaussianNB()
knn1 = KNeighborsClassifier(n_neighbors=5)
```

```python
model1.fit(x1_train,y1_train)
knn1.fit(x1_train,y1_train)
```

### NaiveBayes Accracy Score

```python
y_prd_NB = model1.predict(x1_test)
```

```python
acc_score_NB = accuracy_score(y1_test,y_prd_NB)
```

```python
print(acc_score_NB)
```

```python
confusion_matrix(y1_test,y_prd_NB)
```

```python
print(classification_report(y1_test,y_prd_NB))
```

### KNN Accracy Score

```python
knn1.score(x1_test,y1_test)
```
