```python
#5. Group by make, fuel-type and body-style to calculate the average of city-mpg and highwaympg (miles per gallon
df.groupby(['make','fuel_type','body_style']).agg({'city_mpg':'mean', 
                         'highway_mpg':'mean'})
```
