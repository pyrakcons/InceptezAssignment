### 13.	Perform following aggregation task on below table
### a.	Get the average age group by Regrion
### b.	Get the sum of Age and Validation group by Hair color and then reset the index
### c.	Get the man and the median of Age group by Region


```python
pdlist2 = pd.DataFrame([[25,'Black','SW',1],
                        [30,'Black','NW',1],
                        [45,'Red','NE',1],
                        [35,'Blond','SW',0],
                        [33,'Brown','SW',0],
                        [35,'Brown','NE',0],
                        [25,'Black','SW',0],
                        [52,'Brown','SE',0],
                        [21,'Blond','SW',1],
                        [24,'Blond','NE',0],
                        [48,'Black','SW',1],
                        [22,'Black','SW',0],
                        [27,'Brown','NE',0],
                        [28,'Black','SE',1],
                        [21,'Blond','SE',1],
                        [51,'Black','NW',0],
                        [23,'Brown','SE',1],                        
                       ]
                       ,columns =['AGE','HAIR COLOR','US REGION','VALIDATION'] )
```

```python
pdlist2
```

```python
pdlist2.groupby(['US REGION'])['AGE'].count()
```

```python
pdlist2.groupby(['US REGION'])['AGE'].sum()
```

```python
# M1: Get the average age group by Regrion
pdlist2.groupby(['US REGION'])['AGE'].sum()/pdlist2.groupby(['US REGION'])['AGE'].count()
```

```python
# M2: Get the average age group by Regrion
pdlist2.groupby(['US REGION'])['AGE'].mean().reset_index()
```

```python
 #Get the sum of Age and Validation group by Hair color and then reset the index
pdlist2.groupby(['HAIR COLOR'])['AGE','VALIDATION'].sum().reset_index()
    
```

```python
#c.	Get the man and the median of Age group by Region
pdlist2.groupby(['US REGION'])['AGE'].agg([np.mean, np.median]).reset_index()
```
